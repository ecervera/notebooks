```python
from candle import position, state
```

```python
import cv2
%matplotlib inline
from pylab import imread, imshow, axis
```

```python
img = cv2.imread('img/candle_off.png')
```

```python
found,x,y,w,h = position(img)
if found:
    print('Candle bounding box: (x,y)=(%d,%d), w=%d, h=%d ' % (x,y,w,h))
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
mp_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imshow(mp_img)
a = axis('off')
```

```python
if found:
    lit = state(img,x,y,w,h)
    if lit:
        print('It seems to be ON')
    else:
        print('It seems to be OFF')
```
